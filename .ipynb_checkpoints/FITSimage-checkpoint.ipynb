```python
naxis1 = Ihdu[0].header['naxis1']
naxis2 = Ihdu[0].header['naxis2']
naxis3 = Ihdu[0].header['naxis3']

crpix3 = Ihdu[0].header['crpix3']
cdelt3 = Ihdu[0].header['cdelt3']
crval3 = Ihdu[0].header['crval3']

rest_freq = 215.219259 * 1e9
c         = 299792458

velocity_array = np.zeros( naxis3 )
velocity_array = crval3 + ( np.arange(naxis3) + 1 - crpix3 ) * cdelt3

freq_array = rest_freq * (1 - velocity_array / c)

xpix = 257
ypix = 362
intensity_array = Ihdu[0].data[0][:, ypix, xpix]
#print(intensity_array.shape)
func.plot_spectra(freq_array/1e9, intensity_array, figsize=(9,6) )
#func.plot_spectra(velocity_array/1e3, intensity_array, xlabel='Velocity[km/s]')

#avg_freq, sigma_freq, amp_freq = func.initial_value(freq_array, intensity_array)
#print(avg_freq/1e9, sigma_freq/1e9, amp_freq)

centroid_freq = func.fit_1dgaus( freq_array/1e9, intensity_array, 
                                 plot = True, kind = 'freq', y_err = 0.0019585446,
                                 figsize=(9,6) ) 
centroid_velo = func.fit_1dgaus( freq_array = velocity_array/1e3, intensity_array = intensity_array, 
                                 plot = False, kind = 'velo', xlabel='Velocity[km/s]',
                                 y_err = 0.0019585446)
shift_freq, delta_freq = func.undoppler( freq_array, intensity_array,velocity_array, centroid_freq=centroid_freq, 
                             centroid_velo=centroid_velo, plot=False)

#ideal_int = func.gaussian(freq_array, amp=amp_freq, mean=rest_freq ,sigma=0.01e9)
#func.plot_spectra(freq_array, ideal_int,figsize=(9,6))
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the double Gaussian function
def double_gaussian(x, A1, mu1, sigma1, A2, mu2, sigma2):
    return (A1 * np.exp(-(x - mu1)**2 / (2 * sigma1**2)) +
            A2 * np.exp(-(x - mu2)**2 / (2 * sigma2**2)))

#sigma = np.std(freq_array)

# Initial guess for the parameters [A1, mu1, sigma1, A2, mu2, sigma2]
initial_guess = [0.04, 215.2e9, 0.05e9, 0.06, 215.3e9, 0.01e9]


# Fit the data using curve_fit
popt, pcov = curve_fit(double_gaussian, freq_array, intensity_array, p0=initial_guess,maxfev=50000)

# Plotting the results
plt.scatter(freq_array, intensity_array, label='Data')
plt.plot(freq_array, double_gaussian(freq_array, *popt), label='Fitted Double Gaussian', color='red')
plt.legend()
plt.show()

# Fitted parameters
print("Fitted parameters:", popt)

print('fitted_mean',popt[1]/1e9, popt[4]/1e9)
```
